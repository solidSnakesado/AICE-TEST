AI 모델이 정확하고 신뢰할 수 있는 예측을 하기 위해서는 학습 데이터와 테스트 데이터를 적절히 분할하는 과정이 매우 중요, 이 과정을 통해 모델이 학습 데이터에만 의존하지 않고, 새로운 데이터에서도 잘 작동하도록 만들어야 함, 이번 장에서는 데이터 분할과 교차 검증(K-Fold Cross-Validation), 그리고 과적합 문제를 확인하는 방법에 대해 구체적으로 설명

1. 데이터 분할하기
    - 데이터 분할은 모델이 실제 환경에서 잘 작동하는 지 확인하기 위한 필수 과정, 모델이 학습 데이터에만 맞추어 학습하면 새로운 데이터에서는 예측이 부정확해질수 있음, 이를 방지하기 위해 데이터를 훈련용 데이터와 테스트용 데이터로 나누어 학습과 평가를 분리

    1. 데이터 분할의 기본 개념
        1. 훈련 데이터(Training Data): 모델이 학습하는 데 사용하는 데이터
            - 입력 데이터(X)와 정답 데이터(Y)가 포함
            - 일반적으로 전체 데이ㅌ의 70 ~ 80%를 사용
        
        2. 테스트 데이터(Test Data)
            - 모델의 성능을 평가하기 위해 사용하는 데이터
            - 학습 과정에서는 사용되지 않으며, 모델이 새로운 데이터에서 얼마나 잘 예측하는 지 확인
            - 일반적으로 전체 데이터의 20 ~ 30%를 사용

        3. 쉽게 이해하기: 데이터 분할 과정을 시험 공부로 비유 가능
            - 훈련 데이터: 학생이 시험 공부를 위해 풀어보는 연습문제
            - 테스트 데이터: 실제 시험 문제, 연습문제와 유사하지만 풀어본적 없는 문제, 학생이 연습문제를 완벽히 플었더라고 시험문제를 제대로 풀지 못한다면 공부한 방식에 문제가 있을 수 있음, 마찬가리로 모델도 학습 데이터에서만 잘 작동하고 새로운 데이터에서는 부정확하다면 학습 과정에 문제가 있는 것

2. k-fold 교차 검증
    - 교차 검증은 훈련 데이터를 여러 부분으로 나누어, 각 부분을 번갈아가며 학습과 검증에 사용하는 평가 방법, 단순히 한 번만 훈련용과 테스트용으로 나누는 것이 아니라, 훈련 데이터를 반복적으로 나누어 평가함으로써 모델의 일반화 성능을 더 신뢰성 있게 측정 가능

    1. k-fold 분할
        - K-Fold 교차 검증에서 "K"는 데이터를 나누는 폴드(fold)의 개수를 의미
        - 예를 들어 데이터를 5개의 폴드(K=5)로 나눈다면
        - 데이터를 5개의 그룹으로 분할
            - 각 그룹을 한 번씩 테스트 데이터로 사용하거, 나머지 4개 그룹은 훈련 데이터로 사용
            - 총 5번의 반복을 통해 모델의 성능을 평가

            1. K-Fold 분할 과정
                - 데이터를 K개의 그룹으로 분할
                - 첫 번째 그룹을 테스트 데이터로 나머지 그룹을 훈련 데이터로 사용해 모델을 학습하고 평가
                - 두 번째 그룹을 테스트 데이터로 사용하고 나머지 훈련 데이터로 설정하여 다시 학습과 평가를 진행
                - 이 과정을 K번 반복
                - 모든 평가 결과의 평균을 계산하여 최종 성능을 확인

    2. 교차 검증의 장점
        1. 모델 평가의 신뢰성 증가: 데이터를 여러 방식으로 나누어 평가하므로 모델 성능이 특정 데이터 분할에만 의존하지 않음
        2. 데이터 활용 극대화: 모든 데이터가 훈련 데이터와 테스트 데이터로 사용되므로, 데이터를 효율적으로 사용 가능
        3. 과적합 방지: 여러 번 평가를 진행하기 때문에 모델이 특정 데이터에 과도하게 맞춰지는 문제를 줄일 수 있음

    3. 교차검증 쉽게 이해하기
        - K-Fold 교차 검증을 운동 경기의 리그전으로 생각해 보면
        - 팀 A, B, C, D, E 가 있다고 가정
            - 각 팀이 한 번씩 다른 탐과 경기를 치르는 과정을 교차 검증으로 볼 수 있음
            - 모든 경기가 끝난 후 각 팀의 평균 성적을 계산하는 것이 교차 검증 결과를 얻는 과정과 유사

3. 과적합 확인
    - 과적합이란? 과적합(Overfitting)은 모델이 학습데이터에는 매우 잘 맞지만, 새로운 데이터에서는 성능이 크게 떨어지는 현상을 말함, 학습 데이터에 포함된 노이즈나 세부적인 특징까지 학습한 경우에 발행, 과적합된 모델은 "A 문제집에서는 완벽히 맞추지만, 다른 문제집에서는 잘못된 답을 내는 학생"과 같음
    
    1. 과적합을 확인하는 방법
        1. 훈련 데이타와 테스트 데이터의 성능 비교
            - 훈련 데이터에서의 성능이 높지만 테스트 데이터에서의 성능이 낮다면 과적합을 의심할 수 있음
                - ex> 훈련 데이터의 정확도 95%, 테스트 데이터 정확도 70%

        2. 교차 검증 결과 확인
            1. 모델 단순화
                - 모델이 지나치게 복잡하면 학습 데이터의 세부적은 특성과 노이즈까지 모두 학습하게 되어 과적합이 발생, 이를 방지하려면 모델의 구조를 단순화하거나 불필요한 데이터를 제거해야 함
                    - 쉽게 이해하기: 모델 단순화를 시험 공부로 생각해 보면, 학생이 시험 중비를 하며 책의 모든 단어를 외우려고 한다면 중요한 개념보다는 불필요한 세부사항에만 집중할 가능성이 높음, 시험 성적이 오히려 나빠질 수 있음, 따라서 핵심 개념 위주로 공부하는 것이 더 효율적

                    - 단순화 하는 방법
                        - 필요 없는 특징 제거: 데이터에서 중요하지 않은 요소를 제거
                            - ex> 고객 데이터에서 "전화번호"는 구매 퍄턴과 관련성이 없으므로 제거 가능
                        - 모델 구조 단순화: 딥러닝 모델의 경우 은닉층(hidden layer)이나 뉴런의 개수를 줄임
                            - ex> 복잡한 5층 모델 대산 2~3층 모델로 축소
                        - 간단한 알고리즘 사용: 문제에 따라 복잡한 모델 대산 간단한 모델을 사용
                            - ex> 랜덤 포레스트(Random Forest) 대산 선형 회귀(Linear Rrgression) 사용

            2. 정규화(Regularization)
                - 정규화는 모델이 학습 데이터에 지나치게 맞추지 않도록, 모델의 가중치 크기를 제한하거나 제약을 가하는 기법, 주로 L1 정규화와 L2 정규화를 사용
                    - 쉽게 이해하기: 정규화를 집 청소에 비유해보면, L1 정규화는 집 안에서 불필요한 물건을 골라내 버리는 작업, 집이 더 단순해지고 깔끔해짐, L2 정규화는 집 안 물건들을 잘 정리정돈하여 공간을 효율적으로 사용하는 작업, 물건은 유지하면서도 어지럽지 않게 만듬

                    - L1 정규화(Lasso Regression)
                        - 특징: 중요하지 않은 변수의 가중치를 0으로 만들어 아예 제거
                        - 효과: 모델으 복잡해지는 것을 방지 중요한 변수만 남김

                    - L2 정규화(Ridge Regression)
                        - 특징: 가중치가 과도하게 커지는 것을 억제, 모델이 지나치게 특정 변수에 의존하지 않도록 함
                        - 효과: 모델이 안정적으로 작동하며 과적합 가능성을 줄임

            3. 더 많은 데이터 확보
                - 학습 데이터가 적으면 모델이 충분히 일반화되지 못하고, 학습 데이터에 과도하게 맞추는 과적합 문제가 발생, 더 많은 데이터를 확보하면 모델이 다양한 상황을 학습할 기회를 얻어 과적합을 방지 가능
                    - 쉽게 이해하기: 학생이 시험문제를 위해 단 3개의 문제만 공부했다고 가정, 시험 문제가 100개라면 공부한 3문제와 비슷한 문제가 나왔을 때만 잘 풀 수 있음, 그러나 70~80개의 문제를 공부했다면, 시험에서 다양한 문제를 더 잘 풀 수 있음

                    - 데이터를 늘리는 방법
                        - 새로운 데이터 수집: 예를 들어, 고객 데이터를 더 많이 수집하거나, 설문조사를 통해 추가 데이터를 확보
                        - 데이터 증강(Data Augmentation): 기존 데이터를 변형하여 새로운 데이터를 생성
                            - ex> 이미지를 회전하거나 색상을 변경해 더 많은 이미지 데이터 생성

            4. 드롭아웃(Dropout)
                - 드럽아웃은 딥러닝 모델에서 학습 중 일부 뉴련을 읨의로 비활성화(사용하지 않음)하는 기법, 이렇게 하면 모델이 특정 뉴런에 지나치게 의존하지 않고, 다양한 뉴런 조합으로 학습 가능
                    - 쉽게 이해하기: 드롭아웃을 팁워크 연습으로 생각해보면, 만약 한 팀이 특정 팀원에게만 의존해서 모든 일을 맡긴하뎜, 그 팀원이 없을 때 다른 팀원들이 제대로 일하지 못함, 하지만 드롭아웃처럼 특정 팀원을 무작위로 제외하고 연습한다면, 팀원 모두거 협력하는 연습을 할 수 있고 팀워크 향상에 도움이 됨
                    
                    - 드롭아웃의 특징
                        - 학습 중에만 적용되며, 평가 시에는 모든 뉴런이 활성화
                        - 드롭아웃 비율(ex> 0.2)은 비활성화 뉴런의 비율을 나타냄
