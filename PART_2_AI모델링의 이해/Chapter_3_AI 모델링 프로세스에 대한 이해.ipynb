AI 모델링은 데이터를 활용하여 문제를 해결하는 과정을 체계적으로 수행하는 것, 이 과정에서 데이터를 준비하고 모델을 학습시키며 최종적으로 결과를 평가하는 일련의 단계로 구성. AI 모델링의 성공 여부는 각 단계를 얼마나 철저히 수행하느냐에 따라 크게 달라짐, 이번장에서는 AI 모델링 프로세스의 주요 단계를 하나씩 구제적으로 설명 

1. AI 모델링 프로세스
    - AI 모델링은 크게 다음 6단계로 나눌 수 있음
    
    1. 데이터 확인
        - 데이터 확인은 모델링의 첫 번째 단계로 사용할 데이터의 상태와 특성을 이해하는 과정
        - 왜 중요한가? 데이터를 제대로 이해하지 못하면 잘못된 가정을 하게 되어 모델링 결과가 부정확해질 수 있음, 따라서 데이터를 깊이 분석하고 문제에 접합한 데이터를 사용하는 것이 중요

        1. 확인해야 할 내용
            - 데이터의 형태: 데이터가 표 형태인지, 이미지인지, 텍스트인지 확인
            - 데이터의 크기: 데이터셋의 크기와 샘플 수를 파악
            - 결측치: 누락된 값이 있는지 확인, 결측치를 어떻게 처리할 지 계획
            - 이상치: 비정상적으로 높은 값이나 낮은 값 등 이상치를 탐지
            - 데이터 분포: 각 변수의 값이 어떤 분포를 가지는 지 파악

        2. 데이터 전처리
            - 데이터 전처리는 데이터를 모델이 학습하기 좋은 형태로 가공하는 과정, 데이터 전처리는 모델의 성능에 직접적으로 영향을 미침
            
            1. 주요 작업
                - 결측치 처리: 데이터릐 누락된 값을, 평균, 중앙값 등으로 채우거나, 해당 데이터를 삭제 처리
                - 이상치 처리: 데이터의 분포를 왜곡하는 이상값을 수정하거나 제거
                - 데이터 변환: 데이터를 정규화(Normalization)하거나 표준화(Standardization)하여 모델이 데이터를 고르게 추리할 수 있도록 함
                - 범주형 데이터 인코딩: 텍스트 형식의 데이터를 숫자형으로 변환(ex> 원-핫 인코딩)하여 모델이 이해할 수 있게 처리
            2. 중요성: 데이터 전처리를 통해 노이즈를 줄이고 데이터의 품질을 높여 모델의 성능을 극대화할 수 있음

        3. AI모델 선정
            - AI 모델 선정은 문제를 해결하기에 적합한 알고리즘을 선택하는 단계
            - 어떻게 선택하는 가?

                1. 문제 유형에 따라 모델을 선택
                    - 분류 문제: 이미지 분류, 이메일 스팸 분류 등
                    - 추천 모델: 결정 트리, 랜덤 포레스트, 로지스틱 회귀
                    - 회귀 문제: 주택 가격 예측, 매출 예측 등
                    - 추천 모델: 선형 회귀, 랜덤 포레스트 회귀
                    - 군집화 문제: 고객 세그먼트 분류
                    - 추천 모델: K-Means, 계층적 군집화
                
                2. 데이터 크기와 특성에 따라 모델을 선택
                    - 데이터가 크다면: 딥러닝 모델(CNN, RNN 등)
                    - 데이터가 작다면: 머신러닝 모델(선형 회귀, 결정 트리 등)

                3. 중요성: 적합한 모델을 선택하지 않으면 문제 해결에 필요한 성능을 얻기 어려움, 문제와 데이터에 적합한 모델을 선택하는 것이 핵심

        4. 학습 데이터 분할
            - 학습 데이터 분할은 데이터를 훈련용 데이터와 테스트용 데이터로 나누어 모델을 학습과 평가를 분리하는 과정
            - 왜 나누는 가?
                - 모델이 학습 데이터에 과도하게 적용(과적합)하는 것을 방지하고, 새로운 데이터에 대해 성능을 평가하기 위함

                1. 분할 방식
                    - 훈련 데이터: 모델이 학습하는 데 사용, 전체 데이터의 70~80%를 차지
                    - 테스트 데이터: 모델의 성능을 검증하는 데 사용, 나머자 20~30%를 차지

                2. 추가적으로
                    - 교차 검증(K-Fold Cross Validation): 데이터를 여러 부분으로 나누어 모델의 성능을 안정적으로 평가하는 방법

        5. 모델 학습
            - 모델 학습은 훈련 데이터를 사용해 모델이 데이터를 학습하고 패턴을 이해하는 과정

            1. 학습 방법
                - 데이터를 모델에 입력
                    - 모델은 데이터를 분석하며 입력과 출력 간의 관계를 학습
                    - 학습이 진행될수록 모델으 예측 성능이 점치 향상
            
            2. 학습 중 확인할 사항
                - 손실 함수(Loss Function): 모델의 예측이 실제 정답과 얼마나 차이가 있는 지 측정, 손실값이 낮을수록 모델의 성능이 좋음
                - 최적화 알고리즘(Optimizer): 손실 값을 최소화하기 위해 모댈의 매개변수를 조정
                    - ex> SGD, Adam
                
            3. 중요성: 학습이 잘못 이루어지면 모델이 데이터를 제대로 이해하지 못하고 성능이 저하됨, 학습 속도와 정확성을 적절히 조정하는 것이 중요

        6. 모델 평가
            - 모델 평가 단계에서는 테스트 데이터를 사용햐 모델의 성능을 검증

            1. 평가 방법
                - 분류 모델: 정확도(Accuracy), 정밀도(Precision), 재현율(Recall), F1-score 등을 사용해 평가

                    | 지표 | 정의 | 사용 예 | 
                    | --- | --- | --- |
                    | 정확도 | 전체 데이터 중 예측한 비율 | 데이터가 균형 잡햐 있을 때 적합 | 
                    | 정밀도 | 예측한 긍정 결과 중 실제로 긍정인 비율 | 잘못된 긍정 예측을 줄이는 것이 중요할 때 | 
                    | 재현율 | 실제 긍정 결과 중 모델이 예측한 비율 | 놓치면 안 되는 결과를 강조할 때 | 
                    | FI-score | 정밀도와 재현율의 조화 평균 | 정밀도와 재현율의 균형이 필요할 때 | 

                - 회귀 모델: 평균 제곱 오차(MSE), 평균 절대 오차(MAE), 결정 계수($R^2$)등을 사용해 평가

                    | 지표 | 정의 | 사용 예 | 
                    | --- | --- | --- |
                    | MSE | 예측값과 실제값의 차이를 제곱하여 평균한 값 | 큰 오차를 강조할 때 적합 | 
                    | MAE | 예측값과 실제값의 차이의 절대값 평균 | 전체적인 오차 크기를 파악할 때 적합 | 

                - 왜 중요한가?
                    - 모델 평가를 통해 실제 데이터에서의 성능을 예측할 수 있으며, 모델이 과적합 또는 과소적합인지 확인할 수 있음

        7. AI모델링 프로세스 요약
            1. 데이터 확인: 데이터를 이해하고 준비 상태를 점검
            2. 데이터 전처리: 데이터 품질을 높이고 모델 학습에 접합한 행태로 가공
            3. AI 모델 선정: 문제와 데이터에 가장 적합한 알고리즘을 선택
            4. 학습 데이터 분할: 데이터를 훈련용과 테스트용으로 나누어 학습과 평가를 분리
            5. 모델 학습: 훈련 데이터를 사용해 모델이 패턴을 학습하도록 함
            6. 모델 평가: 테스트 데이터를 사용해 모델 성능을 평가하고 개선 방향을 도출

            - 이처럼 AI 모델링은 각 단계가 유기적으로 연결되어 있으며, 각 단계를 철저히 수행해야 좋은 성과를 얻을 수 있음, 데이터를 다룰 때 실수라도 모델 성능에 큰 영향을 미칠 수 있으므로 꼼꼼한 작업 필요
