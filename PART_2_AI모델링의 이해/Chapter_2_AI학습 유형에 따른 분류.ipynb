AI 모델은 데이터를 학습하고 분석하는 방식에 따라 크게 지도학습(Supervised Learning), 비지도학습(Unsupervised Learning), 그리고 강화학습(Reinforcement Learning)으로 나눌 수 있음, 이번 챕터에서는 이 중 지도학습과 비디조 학습의 원리와 개념을 다룸

1. 지도학습(Supervised Learning) 이해
    1. 정의
        - 지도학습(Supervised Learning)은 입력 데이터와 해당 데이터의 정답(라벨)이 함께 제공되는 학습 방식. 모델은 주어진 데이터를 통해 입력과 정답 간의 관계를 학습하고, 이를 바탕으로 새로운 데이터에 대해 정확한 예측을 수행할 수 있음, 예를 들어 학교에서 선생님이 학생에게 문제와 정답을 함꼐 제공하며 학습시키는 과정과 비슷, 학슴은 문제(입력)를 보고 정답(출력)을 배우며, 이를 바탕으로 시험 문제(새로운 입력)에 대해 답을 작성
    
    2. 작동원리
        1. 데이터 준비
            - 비도학습에서는 데이터를 "입력값(X)"과 출력값(Y)으로 나눔, 예를 들어, "이메일이 스팸인지 아닌지를 뷴류"하는 경우
                - 입력 데이터(X): 이메일 내용 (ex> 단어, 문장)
                - 출력 데이터(Y): 스팸(1) 또는 일반 메일(0)
        2. 모델 학습
            - 준비된 데이터를 사용해 입력(X) -> 출력(Y)의 관계를 학습, 이 과정에서 모델은 데이터를 분석 해 패턴과 규칙을 찾아냄
        3. 예측
            - 학습한 모델을 사용해 새로운 입력 데이터에 대한 결과(출력)를 예측
                - ex> 새로운 이메일이 스팸인지 아닌지 판단
        4. 평가 및 개선
            - 모델의 예측 결과를 실제 정답과 비교하여 성능을 평가, 모델이 부족한 점이 있다면 추가 학습이나 데이터 조정을 통해 개선 진행

    3. 활용 예시
        1. 이미지 분류
            - 입력: 강아지 사진 -> 출력: "강아지" 또는 "고양이"
            - 모델은 수천 장의 동뮬 이미지를 보고 학습하며, 새로운 동물 종류를 예측
        2. 스팸 메일 분류
            - 입력: 이메일 내용 -> 출력: 스팸(1) 또는 일반 메일(0)
            - 모델은 과거 이미일 데이터에서 "스팸 매일의 특징"을 학습
        3. 질병 진단
            - 입력: 환자의 건강 데이터 -> 출력: 특정 질병 여부
            - 모델은 과거 환자의 데이터를 학습해 새로운 환자의 질병 가능성을 예측

2. 비지도학습(Unsupervised Learning) 이해
    1. 정의
        - 비지도 합습(Unsupervised Learning)은 입력 데이터만 제공되고, 이에 해당하는 정답(라벨)이 없는 학습 방식, 모델은 주어진 데이터에서 숨겨진 패턴이나 구조를 학습하고, 데이터를 새로운 방식으로 이해할 수 있도록 도와줌
        - 예를 들어 비지도학습은 "정갑이 없는 퍼즐"과 같음, 학생이 선생님 없이 문제를 스스로 분석하고 답을 찾아내는 과정, 친구 그룹을 성격이나 관심사에 따라 자동으로 분류하는 것과 비슷

    2. 작동원리
        1. 데이터 준비: 비지도학습에서는 입력 데이터(X)만 사용
            - ex> 대형 쇼핑몰의 고객 데이터를 분석하는 경우
                - 입력 데이터(X): 고객의 나이, 구매 횟수, 소비 금액 등.
        2. 모델은 데이터를 분석하여 유사한 데이터끼리 그룹화하거나 특징을 추출
            - 예를 들어, 고객 데이터를 분석하여 다음과 같은 그룹을 발견 가능
                - 그룹 1: 고가 제품을 선호하는 고개
                - 그룹 2: 할인 제품을 주로 구매하는 고객
        3. 결과 활용: 발견한 패턴이나 그룹을 활용해 데이터를 분석하거나 의사결정에 활용
            - ex> 고객 그룹을 나눈 결과를 기반으로 맞춤형 마케팅 전략을 수립

    3. 활용 예시
        1. 고객 세그먼트 분류
            - 입력: 쇼핑몰의 고객 데이터(나이, 소비 금액 등)
                - 모델은 고객을 비슷한 소비 패턴을 가진 그룹으로 나눔
            - 그룹 1: 고가 제품 선호
            - 그룹 2: 할인 제품 선호
        2. 이상치 탐지
            - 입력: 네트워크 트레픽 데이터
                - 모델은 일반적인 트래픽 패턴과 다른 이상한 행동을 탐지
            - 활용: 금융 사기 탐지, 시스템 침입 탐지
        3. 문서 군집화
            - 입력: 수백 개의 문서
                - 모델은 유사한 주제를 가진 문서를 자동으로 그룹화 함
            - 활용: 뉴스 기자 자동 분류

3. 지도학습과 비지도합습의 차이

| 구분 | 지도학습(Supervised Learning) | 비지도학습(Unsupervised Learning) |
| --- | --- | --- |
| 라벨 | 정답(라벨)이 필요함 | 정답(라벨)이 필요 없음 |
| 목표 | 입력과 출력 간 관계 학습 | 데이터 간 패턴이나 구조 발견 |
| 활용 | 예측 문제(분류, 회귀) 해결 | 데이터 분석 및 그룹화, 패턴 탐지 |
| 예시 | 이메일 스팸 분류, 질병 진단 | 고객 세그먼트 분류, 이상치 탐지 |
