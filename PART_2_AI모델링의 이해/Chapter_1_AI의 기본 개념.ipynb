1. 머신러닝의 이해
    - 머신러닝(Machine Learning)은 데이터를 학습하여 일정한 패턴을 발견하고 이를 기반으로 예측하거나 결정을 내리는 기술, 이 과정에서 프로그래밍 언어로 명시적 규칙을 작성하는 대힌, 데이터를 사영하여 컴퓨터가 스스로 학습할 수 있도록 함. 예를 들어, 이메일 서비스에서 스팸 메일을 자동으로 분류해주는 기능을 보면, 스팸 메일과 일반 메일의 차이를 컴퓨터에 설명하기 어려울 수 있음, 그러나 머신러닝을 사용하면 스팸 메일과 일반 메일 데이터를 제공하여 컴퓨터가 스스로 차이점을 학습하도록 할 수 있음, 컴퓨터는 학습한 내용을 바탕으로 새로운 이메일이 들어올 때 스팸인지 아닌지를 예측

   1. 머신러닝의 주요 특징
      - 데이터 중심적 학습: 머신러닝을 데이터를 사용하여 모델을 학습, 이 데이터가 많고 품질이 좋을수록 모델 성능이 향상
      - 일반화 능력: 학습한 데이터 외에 새로운 데이터에서도 높은 성능을 보이는 것을 목표로 함
      - 적응 능력: 환경 변화나 새로운 데이터에 대해 모델을 재학습 가능
   2. 머신러닝의 종류
      - 지도학습(Supervised Learning): 입력 데이터와 함께 정답(라벨)이 제공, 이를 기반으로 모델을 학습
         - ex> 이메일이 스펨인지 아닌지를 뷴류하는 모델
      - 비지도학습(Unsupervised Learning): 정답이 없는 데이터를 기반으로 데이터 내 숨겨진 패턴을 학습
         - ex> 고객 내 세그먼트 군집화
      - 강화학습(Reinfircement Learning): 에이전트가 환경롸 상호작용하며 보상을 최대화하는 방법을 학습
         - ex> 게임 AI, 로봇 제어
   3. 머신러닝의 활용 예시
      - 이미지 분류: 사진 속 인물이나 사물을 구분
      - 추천 시스템: 사용자가 좋아할 만한 영화나 상품 추천
      - 의료 진단: 환자의 데이터를 바탕으로 질병 진단
   4. 머신러닝이 동작하는 과정
      - 데이터 준비: 먼저, 컴퓨터가 학습할 데이터가 필요, 과거 이메일의 데이터를 모아 "스팸"과 "일반"으로 분류
      - 모델 학습: 준비된 데이터를 바탕으로 컴퓨터가 학습, 이 과정에서 데이터의 패턴과 규칙을 찾아냄
      - 예측: 학습한 모델을 사용해 새로운 데이터(ex> 새로운 이메일)가 들어왔을 때 스팸인지 일반 메일인지 예측
      - 평가 및 개선: 예측 결과가 얼마나 정확한지 평가하고, 부족한 점을 개선해 모델을 더 똑똑하게 만듬

2. 딥러닝의 이해
    - 딥러닝(Deep Learning)은 머신러닝의 한 분야로 인공신경망(Artificual Neural Network)이라는 구조를 사용해 데이터를 학습하는 기술. 딥러닝은 특히 방대한 데이터와 복잡한 문제를 처리하는 데 걍력한 성능을 발휘, 딥러닝의 핵심은 사람의 뇌를 모방하여 여러 층(Layer)으로 구성된 신경망을 통해 학습한다는 점, 이 신경망은 데이터를 단계적으로 분석하며, 복잡한 패턴과 구조를 자동으로 이해함

    1. 딥러닝의 주요 특징
        - 계층적 학습 구조
            - 딥러닝 모델은 여러 개의 층을 통해 데이터를 점진적으로 학습
            - 예를 들어
            - 첫 번째 층은 이미지의 가장 기본적인 특징(ex> 선과 곡선)을 학습
            - 두 번째 층은 기본적인 특징을 조합해 복잡한 모양을 학습
            - 마지막 층은 학습한 내용을 바탕으로 이미지를 특정 사물로 분류
        - 대량의 데이터 학습 가능
            - 딥러닝을 대량의 데이터에서 높은 성능을 발휘, 데이터가 많아질수록 성능이 좋아지며, 특히 이미지, 음성, 텍스트 데이터에서 강력
        - 비선현 관계 학습
            - 딥러닝은 단순히 선형 관계뿐만 아니라 복잡하고 비선형적인 데이터 간의 관계를 학습 가능

    2. 딥러닝의 종류
        - 합성곱 신경만(CNN, Convolutional Neural Networks): 이미지 처리에 특화된 딥러닝 모델.
            - ex> 얼굴 인식, 자율주행 차량의 객체 인식
        - 순환 신경망(RNN, Recurrent Neural Network): 시산적 순서가 중요한 데이터를 학습
            - ex> 음성 인식, 텍스트 번역, 주가 예측
        - 생성적 적대 신경망(GAM, Generative Adversarial Network): 새로운 데이터를 생성하는 데 사용
            - ex> 가짜 이미지 생성, 예술작품 스타일 변환
    3. 딥러닝의 활용 예시
        - 이미지 분석: X-ray 이미지를 분석해 질병 진단
        - 음성 인식: 스마트폰이 사용자의 음성을 인식하고 명령 수행
        - 자연어 처리: 챗봇이 사용자의 질문에 적절한 답변 제공
        - 추천 시스템: 넷플릭스에서 사용자가 좋아할 만한 콘텐츠 추천
    4. 딥러닝의 동작 과정
        - 딥러닝은 여러 층(layer)으로 이루어진 신경망을 통해 데이터를 학습, 아래는 기본 동작 과정

        - 데이터 준비: 컴퓨터가 학습할 데이터를 모음, 예를 들어, 수천장의 고양이과 강아지 이미지를 준비해 "고양이", "강아지"로 분류
        - 신경망 설계: 학습에 사용할 신경망의 구조를 설계, 예를 들어
            - 입력층(input layer): 이미지를 받아들이는 층
            - 은닉층(hidden layers): 이미지의 특징을 학습하는 층
            - 출력층(output layer): 학습 결과를 바탕으로 "고양이"인지 "강아지"인지 예측
        - 모델 학습: 데이터를 사용해 신경망에 입력하여 학습, 이 과정에서 신경망은 각 층을 거치며 점차 더 높은 수준의 특징을 학습
        - 예측: 학습한 모델을 사용해 새로운 이미지를 입력했을 때 "고양이"인지 "강아지"인지 예측
        - 평가 및 개선: 예측의 정확성을 평가하고, 성능을 높이기 위해 모델을 조정하거나 추가 학습을 진행
